```python
df = df.rename(columns={"UNIT": "Value"}) # renames a column
df.head()
```

```python
df.isnull().sum() #show the number of rows containing null value
```

```python
# Markdown: In the steps above I completed step B of the lab. I renamed the "UNIT" column into "Value" and I have a lot of missing values, but decided to remove any because I don't know how much overlaps with my data.
```

```python
df = df[['INDICATOR','YEAR','AGE','ESTIMATE',]] # keep specific columns 
df.head()
```

```python
df.describe()
```

```python
df.set_index('YEAR',inplace=True) #define the index
df.head()
```

```python
df = df[(df['ESTIMATE']>=0)] #filter data
df.describe()
```

```python
# Markdown: In the steps above I completed steps C of the lab. I filtered estimate and defined the data using the year column. 
```

```python
df.to_csv('clean_data.csv') #store the cleaned data to local computer
```

```python
df.hist(layout=(8,8), figsize=(10, 5), sharey=False, sharex=False, bins=50) #use multiple charts to show the distribution of all columns
```

```python
sum_estimate_by_year = df.groupby('ESTIMATE')['YEAR'].sum() # calculate the total estimates per year
sum_estimate_by_year.nlargest(10).plot.bar() # use a bar chart to show the results of the top 10 estimates
```
